Also look for independent variables that have high correlations - this is an indication of codependency and could be neutral to bad for your model. A predictor that has nothing to do with danger could be highly correlated with a predictor that does predict danger well, and if you include the bad predictor it will decrease prediction strength. Use this step as a chance to take out codependent factors and reintroduce stronger predictors if you feel the need to."

```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(16,6))

heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')

heatmap.set_title('Correlation Heatmap', fontdict={'fontsize': 12}, pad=12)

plt.savefig('../content/images/heatmap.png', dpi=300, bbox_inches='tight')
```

```python
plt.figure(figsize=(16,6))

mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

heatmap=sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')

heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 18}, pad=16)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor # try classifier next

def get_r2_score(training_data):
    """Returns training score, actual score"""
    drop = 'Danger Rating'
    x = training_data.drop(drop, axis=1)
    y = training_data[drop]

    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)

    rf_regressor = RandomForestRegressor(random_state=42)
    rf_regressor.fit(x_train, y_train)

    y_pred_train = rf_regressor.predict(x_train)
    y_pred = rf_regressor.predict(x_test)


    return r2_score(y_train, y_pred_train), r2_score(y_test, y_pred)
```

```python

# Drop all columns with negative correlation
training, actual = get_r2_score(df.drop(['Sum Max Temp 72hr', 'Max Air Temp 24hr', 'Max Air Temp 72hr'], axis=1))
print('training:', training, '\nactual:\t', actual)
```

```python
# Try dropping every column to get a better score
def find_bad_col(df):
    max_r2 = get_r2_score(df)[1]
    bad_col = None
    for col in df.columns[1:]:
        training, actual = get_r2_score(df.drop([col], axis=1))
        if actual > max_r2:
            max_r2 = actual
            bad_col = col
    print('Increased r2 to:', max_r2, 'by deleting:', bad_col)

find_bad_col(df)

drop_sum_max_temp_72 = df.drop('Sum Max Temp 72hr', axis=1)

find_bad_col(drop_sum_max_temp_72)

drop_was_high_winds_24 = drop_sum_max_temp_72.drop('Was High Winds 24hr', axis=1)

find_bad_col(drop_was_high_winds_24)
```
