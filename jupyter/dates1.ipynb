```python
import pandas as pd
crySS = pd.read_csv('../old_data/seasonal_weather_crystal_summit.csv')
crySS.info()
```

```python
crySS['Date PDT'] = pd.to_datetime(crySS['Date PDT'], format='%b %d %H:%M')
crySS.head()
```

```python
crySS['Date PDT'] = crySS['Date PDT'].dt.strftime('%Y-%m-%d %X')
crySS.head()
crySS.info()
```

```python
 crySS['Date PDT'] = crySS['Date PDT'].str.replace("1900","2023")
crySS.head()
```

```python
crySS['Date PDT'] = pd.to_datetime(crySS['Date PDT']).dt.date

crySS.head()
```
