For every date in danger ratings, get the index of that matching date @ 0:00 in snotel data

```python
for idx, row in march_danger.iterrows():
    date_start = row['Date'].split(',')[0] + ' 0:00'
    idx = morse[(morse == date_start).any(axis=1)].index[0]
    print(morse.iloc[idx])
```

For every date in danger ratings, get the previous 3 days of snotel data

```python
# make sure we can get previous data because df only has one month
march_danger_cut = march_danger.iloc[:-3]

# just do a small sample set for now
march_danger_cut = march_danger_cut.iloc[5:7]

for idx, row in march_danger_cut.iterrows():
    date_start = row['Date'].split(',')[0] + ' 0:00'
    idx = morse[(morse == date_start).any(axis=1)].index[0]
    print(date_start)
    print(morse.iloc[idx+1:idx-1+24*3])
```

For every 3 days of data, aggregate it

```python
# make sure we can get previous data because df only has one month
march_danger_cut = march_danger.iloc[:-3]

# initialize a dataframe for aggregated data
training_data = pd.DataFrame([], columns=morse.columns)
training_data['Danger Rating'] = None

for idx, row in march_danger_cut.iterrows():
    date = row['Date'].split(',')[0]
    date_start = row['Date'].split(',')[0] + ' 0:00'
    idx = morse[(morse == date_start).any(axis=1)].index[0]
    three_days_past = morse.iloc[idx+1:idx-1+24*3]
    # aggregate the data
    agg_past = three_days_past.drop('Date PDT', axis=1).mean()
    agg_past['Date PDT'] = date
    agg_past['Danger Rating'] = row['Danger Rating']
    # put it into the dataframe
    training_data.loc[-1] = agg_past
    training_data.index = training_data.index+1

# todo: validate to make sure I did this right
training_data
```

Map danger ratings to numbers

```python
danger_number = { 'NO RATING': 0, 'LOW': 1, 'MODERATE': 2, 'CONSIDERABLE': 3, 'HIGH': 4, 'EXTREME': 5 }

training_data['Danger Rating'] = training_data['Danger Rating'].map(danger_number)
training_data.head()
```

That is pretty darn sweet! Now lets try doing some training.

Following [this](https://machinelearningknowledge.ai/random-forest-regression-in-python-sklearn-with-example/) example

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor

# date isn't a number, and isn't relevant (yet)
training_data = training_data.drop('Date PDT', axis=1)

drop = 'Danger Rating'
x = training_data.drop(drop, axis=1)
y = training_data[drop]

x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

rf_regressor = RandomForestRegressor(random_state=0)

rf_regressor.fit(x_train, y_train)

y_pred_train = rf_regressor.predict(x_train)
r2_score(y_train, y_pred_train)
```

```python
y_pred = rf_regressor.predict(x_test)
r2_score(y_test, y_pred)
```
